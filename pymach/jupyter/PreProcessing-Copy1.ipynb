```python
from sklearn.model_selection import KFold
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import KNeighborsRegressor

mlp = MLPRegressor(max_iter=1000)
knn =  KNeighborsRegressor(n_neighbors=5, n_jobs=-1)
kfold = KFold(n_splits=10, random_state=7)
```

```python
import pandas as pd
data = pd.read_csv("titanic_data.csv")
data.head()
print(type(data))
X = data[data.columns[0:4]]
y = data[data.columns[4]]
print(X)
```

```python
from sklearn import preprocessing
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
X_1 = X.select_dtypes(include=numerics)
X_2 = X.select_dtypes(include=[object])
le = preprocessing.LabelEncoder()
X_2 = X_2.apply(le.fit_transform)
X = pd.concat([X_1,X_2],axis=1)
print(X)
print(X.dtypes)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler, RobustScaler
robust_scaler = RobustScaler()
rs = robust_scaler.fit_transform(X_train)
rs
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3, svd_solver='randomized', whiten=True)
pa = pca.fit_transform(rs)
pa
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
kbest = SelectKBest(score_func=f_regression, k=3)
k = kbest.fit_transform(pa,y_train)
print(k)
```

```python
from sklearn.model_selection import cross_val_score
m = cross_val_score(mlp, k, y_train, cv=kfold, scoring='neg_mean_squared_error')
```
