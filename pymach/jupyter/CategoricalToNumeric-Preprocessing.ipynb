```python
from sklearn.model_selection import KFold
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import KNeighborsRegressor

mlp = MLPRegressor(max_iter=1000)
knn =  KNeighborsRegressor(n_neighbors=5, n_jobs=-1)
kfold = KFold(n_splits=10, random_state=7)
```

```python
import pandas as pd
data = pd.read_csv("automovil.csv")
#print(data.columns.values)
count = len(data.columns.values) - 1
string = data.columns.values[count]
data.rename(columns={string:'class'}, inplace=True)
data.describe().round(3)
```

```python
X = data.loc[:, data.columns != 'class']
y = data.loc[:, data.columns == 'class']
y.head()
```

```python
from sklearn import preprocessing
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
X_1 = X.select_dtypes(include=numerics)
X_1.head()
```

```python
X_2 = X.select_dtypes(include=[object])
X_2.head()
```

```python
le = preprocessing.LabelEncoder()
X_2 = X_2.apply(le.fit_transform)
X = pd.concat([X_1,X_2],axis=1)
X.dtypes
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler, RobustScaler
robust_scaler = RobustScaler()
rs = robust_scaler.fit_transform(X_train)
rs
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10, svd_solver='randomized', whiten=True)
pa = pca.fit_transform(rs)
pa
```
